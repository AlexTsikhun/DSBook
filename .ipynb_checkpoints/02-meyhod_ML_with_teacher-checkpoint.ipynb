```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
```

```python
print("Прогнози на тестовій виборці: ", clf.predict(X_test))
```

```python
print("Правильнітсь на тестовому наборі: ", clf.score(X_test, y_test))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # create clf
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X_train, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:, 1], y, ax=ax)
    ax.set_title("Кількість сусідів: {}".format(n_neighbors))
    ax.set_xlabel("0")
    ax.set_xlabel("1")
axes[0].legend(loc=3)
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
# try n_neighbors = 1 to 10
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    # make model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # Write trueful on train data
    training_accuracy.append(clf.score(X_train, y_train))
    # Write trueful on test's data
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="Правильність на трейні")
plt.plot(neighbors_settings, test_accuracy, label="Правильність на тесті")
plt.ylabel("Правильність")
plt.xlabel("Кількість сусідів")
plt.legend()
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)
# розбиваєм набір даних wave на тест і трейн
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
# робим екземпляр моделі і встановлюєм кількість сусідів 3
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
print(reg.predict(X_test))
print('\n', reg.score(X_test, y_test))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# робим 1000 різних точок, розділяєм рівномірно між-3 і 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title("{} neighbor(s)\n train score:{:.2f} test score: {:.2f}".format(n_neighbors,
                 reg.score(X_train, y_train),
                 reg.score(X_test, y_test)))
    ax.set_xlabel("Ознака")
    ax.set_ylabel("Цільова змінна")

axes[0].legend(["Прогнози", "Навчальні дані", "Тестові дані"], loc='best')
```

# Лассо

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)

print("train: ", lasso.score(X_train, y_train))
print("test: ", lasso.score(X_test, y_test))
print("Кількість використаних ознак: ", np.sum(lasso.coef_!=0))
```

```python
# мы увеличиваем значение "max_iter",
# иначе модель выдаст предупреждение, что нужно увеличить max_iter.
lasso001 = Lasso(alpha=0.01, max_iter=10000).fit(X_train, y_train)

print("train: ", lasso001.score(X_train, y_train))
print("test: ", lasso001.score(X_test, y_test))
print("Кількість використаних ознак: ", np.sum(lasso001.coef_!=0))
```

```python
lasso0001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)

print("train: ", lasso0001.score(X_train, y_train))
print("test: ", lasso0001.score(X_test, y_test))
print("Кількість використаних ознак: ", np.sum(lasso0001.coef_!=0))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(lasso0001.coef_, 'v', label="Lasso alpha=0.0001")

plt.plot(ridge01.coef_, 'o', label="Ridge regresion alpha=0.1") # best decision for Ridge
plt.legend(ncol=2)
plt.ylim(-25, 25)
plt.xlabel("Індекс коеф")
plt.ylabel("Оцінка коеф")
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)
logisticRegres = LogisticRegression().fit(X_train, y_train)
print(logisticRegres.score(X_train, y_train))
print(logisticRegres.score(X_test, y_test))
```

```python
logreg100 = LogisticRegression(C=100).fit(X_train, y_train)
print("{:.3f}".format(logreg100.score(X_train, y_train)))
print("{:.3f}".format(logreg100.score(X_test, y_test)))
```

```python
logreg001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.3f}".format(logreg001.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.3f}".format(logreg001.score(X_test, y_test)))
```

```python
plt.plot(logisticRegres.coef_.T, 'o', label='C=1')
plt.plot(logreg100.coef_.T, '^', label='C=100')
plt.plot(logreg001.coef_.T, 'v', label='C=0.01')
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0,0, cancer.data.shape[1])
plt.ylim(-5, 5)
plt.xlabel("Індекс коефіцієнта")
plt.ylabel("Оцінка коефіцієнта")
plt.legend()
```

```python
for C, marker, in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(C=C, penalty='l1', solver='liblinear').fit(X_train, y_train)
    print("Правильность на обучении для логрегрессии l1 с C={:.3f}: {:.2f}".format(C,
                                                                                   lr_l1.score(X_train, y_train)))
    print("Правильность на тесте для логрегрессии l1 с C={:.3f}: {:.2f}".format(C,
                                                                                lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.xlabel("Index koef")
plt.ylabel("Estimator koef")
plt.ylim(-5, 5)
plt.legend()
```

```python
linearSVM = LinearSVC().fit(X, y)
print("Form koef: ", linearSVM.coef_)
print("Form const: ", linearSVM.intercept_)
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linearSVM.coef_, linearSVM.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept)/coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("0")
plt.ylabel("01")
plt.title("Границы принятия решений, полученные с помощью трех бинарных классификаторов\n в рамках подхода «один против остальных")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line of class 0', 'Line of class 1', 'Line of class 2'], loc=(1.01, 0.3))
```

```python
import pandas as pd
ram_prices = pd.read_csv("C:/Users/lenovo/Desktop/DSBook/ram_price.csv")

plt.semilogy(ram_prices.date, ram_prices.price)
plt.xlabel("Year")
plt.ylabel("Cost $/Mbait")
plt.title("Историческое развитие цен на RAM \nпо логарифмической шкале")
```

```python
ram_prices.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
# використовуємо історичні дані, для прогнозу цін після 200року
data_train = ram_prices[ram_prices.date < 2000]
data_test = ram_prices[ram_prices.date >= 2000]

# прогнозуєм ціни по датах
X_train = data_train.date[:, np.newaxis]
# викор ло
y_train = np.log(data_train.price)

tree = DecisionTreeRegressor().fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)

# прогнозуєм по усіх даних
X_all = ram_prices.date[:, np.newaxis]

pred_tree = tree.predict(X_all)
pred_lr = linear_reg.predict(X_all)

# експоненціюєм, щоб переробити голарифмічне 
price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)
```

```python
X, y = make_blobs(centers=4, random_state=8)
y = y % 2
mglearn.discrete_scatter(X[:, 0], X[:,1], y)
plt.xlabel("0")
plt.ylabel("1")
plt.title("""Набор данных с двухклассовой классификацией,
в котором классы линейно неразделимы""")
```

```python
from sklearn.svm import LinearSVC
linSVM = LinearSVC().fit(X, y)

mglearn.plots.plot_2d_separator(linSVM, X)
mglearn.discrete_scatter(X[:,0], X[:, 1], y)
plt.xlabel('0')
plt.ylabel('1')
```

```python
# добавлєм другу ознаку в кдвадраті
X_new = np.hstack([X, X[:, 1:]**2])

from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
# візуалізація в 3D
ax = Axes3D(figure, elev=-152, azim=-26)
# спершу розміщуєм на графіку всі точки з y==0, потім з y==1
mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2],
           c='b', cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2],
          c='r', marker='^', cmap=mglearn.cm2, s=60)
ax.set_xlabel('0')
ax.set_ylabel('1')
ax.set_zlabel('1 ** 2')
```

Вводим лінійну модель, щоб розділити 2а класи

```python
linSVM_3d = LinearSVC().fit(X_new, y)
coef, intercept = linSVM_3d.coef_.ravel(), linSVM_3d.intercept_

# показати границю прийняття рішень лінійної моделі
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_new[:, 0].min() -2 , X_new[:, 0].max()+2, 50)
yy = np.linspace(X_new[:, 1].min() -2, X_new[:, 1].max()+2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2],
          c='b', cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2],
          c='r', marker='^', cmap=mglearn.cm2, s=60)

ax.set_xlabel('0')
ax.set_ylabel("1")
ax.set_zlabel('1**2')
```

Фактически модель линейного SVM как функция исходных 
признаков не является больше линейной. Это не линия, а скорее эллипс

```python
ZZ = YY ** 2
dec = linSVM_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()],
            cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("0")
plt.ylabel("1")
```

## Розуміння принципів роботи SVM

```python
from sklearn.svm import SVC
X, y = mglearn.tools.make_handcrafted_dataset()
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
# розміщуєм на графіку опорні вектори
sv = svm.support_vectors_
# мітки класів опорних векторів визначаються знаком дуальних коефіцієнтів
sv_labels = svm.dual_coef_.ravel()>0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels,
                        s=15, markeredgewidth=3)
plt.xlabel("0")
plt.ylabel("1")
plt.title("Граница принятия решений и опорные векторы,\n найденные SVM с помощью ядра RBF")
```
